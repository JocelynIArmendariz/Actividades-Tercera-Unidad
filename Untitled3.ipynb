```python
import matplotlib.pyplot as plt

# Datos
lanzamientos = ['SSS', 'SSS', 'SSS', 'FSF', 'FFS', 'SSS', 'FFF', 'FFS']

# Conteo de éxitos
exitos_por_lanzamiento = [tiro.count('S') for tiro in lanzamientos]

# Conteo de fallas
fallas_por_lanzamiento = [tiro.count('F') for tiro in lanzamientos]

# Probabilidad de éxito y falla
probabilidad_exito = sum(exitos_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))
probabilidad_falla = sum(fallas_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))

# Crear gráfico de probabilidad de éxito
plt.bar(['Éxito'], [probabilidad_exito], color='lavender', label='Probabilidad de Éxito')

# Crear gráfico de probabilidad de falla
plt.bar(['Falla'], [probabilidad_falla], color='pink', label='Probabilidad de Falla')

plt.ylabel('Probabilidad')
plt.title('Probabilidad de Éxito y Falla en los Lanzamientos')
plt.legend()
plt.grid(True ,color='violet')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
lanzamientos = ['GGG', 'BBB', 'GBG', 'GBB', 'GGG', 'BBB', 'GGG', 'BGB', 'GGB', 'BBG']

# Conteo de niñas
niñas_por_lanzamiento = [tiro.count('G') for tiro in lanzamientos]

# Conteo de niños
niños_por_lanzamiento = [tiro.count('B') for tiro in lanzamientos]

# Probabilidad de niñas y niños
probabilidad_niñas = sum(niñas_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))
probabilidad_niños = sum(niños_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))

# Crear gráfico de probabilidad de niñas
plt.bar(['Niñas'], [probabilidad_niñas], color='lightpink', label='Probabilidad de Niñas')

# Crear gráfico de probabilidad de niños
plt.bar(['Niños'], [probabilidad_niños], color='lightblue', label='Probabilidad de Niños')

plt.ylabel('Probabilidad')
plt.title('Probabilidad de Niñas y Niños en los Lanzamientos')
plt.legend()
plt.grid(True,  color='mediumpurple')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de los lanzamientos
lanzamientos = ['ARA', 'RRA', 'AAR', 'RRR', 'RRA', 'RAR', 'RRR', 'AAA']

# Conteo de veces que el pez es rojo
pez_rojo_por_lanzamiento = [tiro.count('R') for tiro in lanzamientos]

# Número máximo de veces que el pez es rojo
max_vez_rojo = max(pez_rojo_por_lanzamiento)

# Frecuencia relativa de veces que el pez es rojo
frecuencia_relativa = [pez_rojo_por_lanzamiento.count(i) / len(pez_rojo_por_lanzamiento) for i in range(max_vez_rojo + 1)]

# Crear gráfico de barras de la frecuencia relativa
plt.bar(range(max_vez_rojo + 1), frecuencia_relativa, align='center', color='lightpink')
plt.xlabel('Número de Veces que el Pez es Rojo')
plt.ylabel('Frecuencia Relativa')
plt.title('Frecuencia Relativa de Veces que el Pez es Rojo')
plt.xticks(range(max_vez_rojo + 1))
plt.grid(True , color='palevioletred')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos d
lanzamientos = ['AAA', 'AAA', 'ASA', 'SSA', 'AAS', 'SAS', 'AAS', 'ASA', 'AAA', 'SSS']

# Conteo de veces que la lagartija está despierta
lagartija_despierta_por_lanzamiento = [tiro.count('A') for tiro in lanzamientos]

# Conteo de veces que la lagartija está dormida
lagartija_dormida_por_lanzamiento = [tiro.count('S') for tiro in lanzamientos]

# Probabilidad de la lagartija despierta y dormida
probabilidad_despierta = sum(lagartija_despierta_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))
probabilidad_dormida = sum(lagartija_dormida_por_lanzamiento) / (len(lanzamientos) * len(lanzamientos[0]))

# Crear gráfico de barras de la probabilidad de la lagartija despierta o dormida
plt.bar(['Despierta'], [probabilidad_despierta], color='khaki', label='Probabilidad de Despierta')
plt.bar(['Dormida'], [probabilidad_dormida], color='plum', label='Probabilidad de Dormida')

plt.ylabel('Probabilidad')
plt.title('Probabilidad de la Lagartija Despierta o Dormida durante el Día')
plt.legend()
plt.grid(True , color='thistle')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de los platos en la mesa
platos_en_mesa = ['AMAA', 'AMMA', 'AMMM', 'AAMA', 'AMAM', 'AAAM', 'AAMA', 'AAMM', 'MMAM', 'AMAM']

# Conteo de aperitivos por mesa
aperitivos_por_mesa = [mesa.count('A') for mesa in platos_en_mesa]

# Número máximo de aperitivos
max_aperitivos = max(aperitivos_por_mesa)

# Frecuencia relativa de aperitivos por mesa
frecuencia_relativa = [aperitivos_por_mesa.count(i) / len(aperitivos_por_mesa) for i in range(max_aperitivos + 1)]

# Crear gráfico de barras de la frecuencia relativa
plt.bar(range(max_aperitivos + 1), frecuencia_relativa, align='center', color='lightgreen')
plt.xlabel('Número de Aperitivos en la Mesa')
plt.ylabel('Frecuencia Relativa')
plt.title('Frecuencia Relativa de Aperitivos en una Mesa')
plt.xticks(range(max_aperitivos + 1))
plt.grid(True  , color='darkseagreen')
plt.show()
```
